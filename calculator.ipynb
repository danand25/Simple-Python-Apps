```python
import tkinter as tk

# create window
root = tk.Tk()
root.title("Simple Calculator")

# update entries
equation = tk.StringVar()

# create display box
'''display = tk.Entry(root, width=25, state="readonly", justify="right")'''
display = tk.Entry(root, width=25, textvariable=equation, justify="right")
display.pack(pady=10,padx=10)

# create functions
def show_number(num): 
    # display number
    equation.set(equation.get() + str(num))
def add_decimal():
    # display decimal point
     equation.set(equation.get() + ".")
def add_operator(op):
    # display operators +, -, *, /, ^
     equation.set(equation.get() + str(op))
def calculate_result():
    # compute result
    expr = equation.get().replace("^", "**")  # replace ^ with Python's exponent operator
    try:
        result = str(eval(expr))  # carefully: eval works for basic arithmetic
        equation.set(result)
    except:
        equation.set("Error")
def backspace():
    # delete last character
    equation.set(equation.get()[:-1])
def clear_all():
    # clear display
    equation.set("")

# create grid
btn_frame = tk.Frame(root)
btn_frame.pack(pady=5)

# buttons 

#create 0 button
btn_0 = tk.Button(btn_frame, text="0", command=lambda: show_number("0"))
btn_0.grid(row=5,column=1,pady=2,padx=5)

#create 1 button
btn_1 = tk.Button(btn_frame, text="1", command=lambda: show_number("1"))
btn_1.grid(row=4,column=1,pady=2,padx=5)

#create 2 button
btn_2 = tk.Button(btn_frame, text="2", command=lambda: show_number("2"))
btn_2.grid(row=4,column=2,pady=2,padx=5)

#create 3 button
btn_3 = tk.Button(btn_frame, text="3", command=lambda: show_number("3"))
btn_3.grid(row=4,column=3,pady=2,padx=5)

#create 4 button
btn_4 = tk.Button(btn_frame, text="4", command=lambda: show_number("4"))
btn_4.grid(row=3,column=1,pady=2,padx=5)

#create 5 button
btn_5 = tk.Button(btn_frame, text="5", command=lambda: show_number("5"))
btn_5.grid(row=3,column=2,pady=2,padx=5)

#create 6 button
btn_6 = tk.Button(btn_frame, text="6", command=lambda: show_number("6"))
btn_6.grid(row=3,column=3,pady=2,padx=5)

#create 7 button
btn_7 = tk.Button(btn_frame, text="7", command=lambda: show_number("7"))
btn_7.grid(row=2,column=1,pady=2,padx=5)

#create 8 button
btn_8 = tk.Button(btn_frame, text="8", command=lambda: show_number("8"))
btn_8.grid(row=2,column=2,pady=2,padx=5)

#create 9 button
btn_9 = tk.Button(btn_frame, text="9", command=lambda: show_number("9"))
btn_9.grid(row=2,column=3,pady=2,padx=5)

#create + button
btn_plus = tk.Button(btn_frame, text="+", command=lambda: add_operator("+"))
btn_plus.grid(row=1,column=4,pady=2,padx=5)

#create - button
btn_minus = tk.Button(btn_frame, text="-", command=lambda: add_operator("-"))
btn_minus.grid(row=2,column=4,pady=2,padx=5)

#create * button
btn_asterik = tk.Button(btn_frame, text="*", command=lambda: add_operator("*"))
btn_asterik.grid(row=3,column=4,pady=2,padx=5)

#create / button
btn_div = tk.Button(btn_frame, text="/", command=lambda: add_operator("/"))
btn_div.grid(row=4,column=4,pady=2,padx=5)

#create ^ button
btn_exp = tk.Button(btn_frame, text="^", command=lambda: add_operator("^"))
btn_exp.grid(row=1,column=3,pady=2,padx=5)

#create . button
btn_dec = tk.Button(btn_frame, text=".", command=add_decimal)
btn_dec.grid(row=5,column=2,pady=2,padx=5)

#create = button
btn_equal = tk.Button(btn_frame, text="=", command=calculate_result)
btn_equal.grid(row=5,column=4,pady=2,padx=5)

#create del button
btn_del = tk.Button(btn_frame, text="Del", command=backspace)
btn_del.grid(row=1,column=2,pady=2,padx=5)

#create AC button
btn_clear = tk.Button(btn_frame, text="AC", command=clear_all)
btn_clear.grid(row=1,column=1,pady=2,padx=5)

# run calculator
root.mainloop()

```
