```python
import tkinter as tk 

# create window
root = tk.Tk()
root.title("Conversion Calculator")

equation = tk.StringVar()

# create display box
display = tk.Entry(root, width=25, textvariable=equation, justify="left")
display.pack(padx=10, pady=10)

# create functions
def show_number(num):
    equation.set(equation.get() + str(num))
def add_decimal():
    equation.set(equation.get() + ".")
def backspace():
    equation.set(equation.get()[:-1])
def clear_all():
    equation.set("")
def km_to_miles():
    try: 
        result = float(equation.get())*0.621371
        equation.set(result)
    except: 
        equation.set("Error")
def miles_to_km():
    try: 
        result = float(equation.get())*1.609
        equation.set(result)
    except:
        equation.set("Error")
def kg_to_lobes():
    try: 
        result = float(equation.get())*2.205
        equation.set(result)
    except:
        equation.set("Error")
def lobes_to_kg():
    try: 
        result = float(equation.get())*0.454
        equation.set(result)
    except: 
        equation.set("Error")
def F_to_C():
    try: 
        result = ((float(equation.get())-32)*(5/9))
        equation.set(result)
    except:
        equation.set("Error")
def C_to_F():
    try:
        result = ((9/5)*(float(equation.get()))+32)
        equation.set(result)
    except:
        equation.set("Error")

# create buttons 
btn_function = tk.Frame(root)
btn_function.pack()

# 0 button
btn_0 = tk.Button(btn_function, text = "0", command=lambda: show_number("0"))
btn_0.grid(row=5,column=1,pady=2,padx=5)

# 1 button
btn_1 = tk.Button(btn_function, text = "1", command=lambda: show_number("1"))
btn_1.grid(row=4,column=1,pady=2,padx=5)

# 2 button
btn_2 = tk.Button(btn_function, text = "2", command=lambda: show_number("2"))
btn_2.grid(row=4,column=2,pady=2,padx=5)

# 3 button
btn_3 = tk.Button(btn_function, text = "3", command=lambda: show_number("3"))
btn_3.grid(row=4,column=3,pady=2,padx=5)

# 4 button
btn_4 = tk.Button(btn_function, text = "4", command=lambda: show_number("4"))
btn_4.grid(row=3,column=1,pady=2,padx=5)

# 5 button
btn_5 = tk.Button(btn_function, text = "5", command=lambda: show_number("5"))
btn_5.grid(row=3,column=2,pady=2,padx=5)

# 6 button
btn_6 = tk.Button(btn_function, text = "6", command=lambda: show_number("6"))
btn_6.grid(row=3,column=3,pady=2,padx=5)

# 7 button
btn_7 = tk.Button(btn_function, text = "7", command=lambda: show_number("7"))
btn_7.grid(row=2,column=1,pady=2,padx=5)

# 8 button
btn_8 = tk.Button(btn_function, text = "8", command=lambda: show_number("8"))
btn_8.grid(row=2,column=2,pady=2,padx=5)

# 9 button
btn_9 = tk.Button(btn_function, text = "9", command=lambda: show_number("9"))
btn_9.grid(row=2,column=3,pady=2,padx=5)

# . button
btn_dec = tk.Button(btn_function, text = ".", command=lambda: add_decimal())
btn_dec.grid(row=5,column=3,pady=2,padx=5)

# del button
btn_del = tk.Button(btn_function, text = "Del", command=lambda: backspace())
btn_del.grid(row=1,column=2,pady=2,padx=5)

# clear button
btn_clear = tk.Button(btn_function, text = "AC", command=lambda: clear_all())
btn_clear.grid(row=1,column=1,pady=2,padx=5)

# kilometers to miles
btn_km_to_miles = tk.Button(btn_function, text = "km to miles", command=lambda: km_to_miles())
btn_km_to_miles.grid(row=2,column=4,pady=2,padx=5)

# miles to kilometers
btn_miles_to_km = tk.Button(btn_function, text = "miles to km", command=lambda: miles_to_km())
btn_miles_to_km.grid(row=3,column=4,pady=2,padx=5)

# kilograms to lobes
btn_kg_to_lobes = tk.Button(btn_function, text = "kg to lobes", command=lambda: kg_to_lobes())
btn_kg_to_lobes.grid(row=4,column=4,pady=2,padx=5)

# lobes to kilograms
btn_lobes_to_kg = tk.Button(btn_function, text = "lobes_to_kg", command=lambda: lobes_to_kg())
btn_lobes_to_kg.grid(row=5,column=4,pady=2,padx=5)

# fahrenheit to celsius
btn_F_to_C = tk.Button(btn_function, text = "F to C", command=lambda: F_to_C())
btn_F_to_C.grid(row=1,column=4,pady=2,padx=5)

# celsius to fahrenheit
btn_C_to_F = tk.Button(btn_function, text = "C to F", command=lambda: C_to_F())
btn_C_to_F.grid(row=1,column=3,pady=2,padx=5)

root.mainloop()
```
